## 연령대별 서비스 히트맵 

```python
data = pd.read_excel('./zvalue_services.xlsx', index_col=0)

for i in range(2, 8):
    df = pd.DataFrame()

    df = data[data['연령대'] == i * 10]

    df = df.drop(columns = ['연령대'])

    correlation_matrix = df.corr()

    plt.figure(figsize=(10, 8))
    sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
    plt.title('{}0대 서비스별 상관관계 히트맵'.format(i))
    plt.show()
```

## 서비스별로 자치구 이용 분포

```python
data = pd.read_excel('./zvalue_services.xlsx', index_col=0)

df = pd.DataFrame()

df = data.groupby('자치구')['게임 서비스 사용일수', '금융 서비스 사용일수', '쇼핑 서비스 사용일수', '동영상/방송 서비스 사용일수', '유튜브 사용일수', '넷플릭스 사용일수', '배달 서비스 사용일수'].mean()

df
```

```python
f, axes = plt.subplots(4, 2)
f.set_size_inches((12, 16))
plt.subplots_adjust(wspace = 0.15, hspace = 0.15)

# figure 전체 제목
f.suptitle('서비스별 자치구 이용 분포', fontsize = 15)


def graph(i):
    plt.subplot(3, 2, i)
    service = df[df.columns[i]] 
    particular_service = service.sort_values()

    particular_service.plot(kind='barh', color='cadetblue')
    plt.title(df.columns[i])
    plt.ylabel('')

for i in range(1, 7):
    graph(i)
```
